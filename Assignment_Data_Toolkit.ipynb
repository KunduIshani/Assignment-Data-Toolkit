```python
###1. Three Methods for Creating Identical 2D Arrays in NumPy
```

```python
import numpy as np

# Method 1: Using np.full
array1 = np.full((2, 2), 7)
print(array1)

# Method 2: Using np.ones multiplied by a constant
array2 = np.ones((2, 2)) * 7
print(array2)

# Method 3: Using np.tile
array3 = np.tile(7, (2, 2))
print(array3)
```

```python
###14. Create a Time-Series Dataset and Plot Temperature and Humidity
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create time-series dataset
date_rng = pd.date_range(start='2024-01-01', end='2024-12-31', freq='D')
temperature = np.random.randint(20, 35, size=len(date_rng))
humidity = np.random.randint(30, 70, size=len(date_rng))

df = pd.DataFrame({'Date': date_rng, 'Temperature': temperature, 'Humidity': humidity})

# Plot temperature and humidity on different y-axes
fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.plot(df['Date'], df['Temperature'], 'g-')
ax2.plot(df['Date'], df['Humidity'], 'b-')

ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature', color='g')
ax2.set_ylabel('Humidity', color='b')

plt.title('Temperature and Humidity Over Time')
plt.show()
```

```python
###15. Histogram with PDF Overlay Using Matplotlib
```

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats

# Generate normal distribution data
data = np.random.normal(loc=0, scale=1, size=1000)

# Plot histogram
plt.hist(data, bins=30, density=True, alpha=0.6, color='g')

# Overlay PDF
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = stats.norm.pdf(x, 0, 1)
plt.plot(x, p, 'k', linewidth=2)

plt.xlabel('Value')
plt.ylabel('Frequency/Probability')
plt.title('Histogram with PDF Overlay')
plt.show()
```

```python
###16.set the title of the plot as ' Histogram with PDF Overlay'
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Generate data from a normal distribution
data = np.random.normal(0, 1, 1000)

# Plot the histogram with density=True to normalize it
plt.hist(data, bins=30, density=True, alpha=0.6, color='g')

# Get the limits of the x-axis for plotting the PDF
xmin, xmax = plt.xlim()

# Generate 100 points on the x-axis between xmin and xmax
x = np.linspace(xmin, xmax, 100)

# Generate the PDF for a normal distribution with mean=0 and std=1
p = stats.norm.pdf(x, 0, 1)

# Plot the PDF line on the same plot as the histogram
plt.plot(x, p, 'k', linewidth=2)

# Set the title of the plot
plt.title('Histogram with PDF Overlay')

# Label the axes
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')

# Show the plot
plt.show()
```

```python
###17.create a seabird scatter plot of two random arrays,, colour points based on their positive relative to the origin (quadrants), add a legend, label the axes, and set the title as "Quadrant-wise scatter plot."
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generate random x and y data between -1 and 1
x = np.random.uniform(-1, 1, 100)
y = np.random.uniform(-1, 1, 100)

# Determine the quadrant for each point
quadrants = ['Q1' if xi > 0 and yi > 0 else
             'Q2' if xi < 0 and yi > 0 else
             'Q3' if xi < 0 and yi < 0 else
             'Q4'
             for xi, yi in zip(x, y)]

# Create a scatter plot with different colors for each quadrant
plt.figure(figsize=(8, 6))
sns.scatterplot(x=x, y=y, hue=quadrants, palette='Set1', s=100)

# Label the axes
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Set the title
plt.title('Quadrant-wise Scatter Plot')

# Show the plot
plt.legend(title='Quadrants')
plt.show()
```

```python
###18. with bokeh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'sine wave function'.
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

# Display the output in the notebook
output_notebook()

# Step 1: Create the sine wave data
x = np.linspace(0, 4 * np.pi, 100)  # x values from 0 to 4*pi
y = np.sin(x)  # sine of the x values

# Step 2: Create the Bokeh figure
p = figure(title="Sine Wave Function", x_axis_label='X-axis', y_axis_label='Y-axis')

# Step 3: Plot the sine wave line
p.line(x, y, legend_label="Sine Wave", line_width=2)

# Step 4: Add grid lines (this happens automatically, but we can adjust them if needed)
p.grid.grid_line_color = "gray"

# Step 5: Display the plot
show(p)
```

```python
###19.using bokeh, generate a bar chart of randomly generated categorical data, cool bars based on their values, add hover tooltips to display exact values, label the axes, and set the title as 'random categorical bar chart'.
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.transform import factor_cmap
from bokeh.palettes import Viridis256
import numpy as np

# Show Bokeh plots in the notebook (if using a notebook)
output_notebook()

# Step 1: Generate random categorical data
categories = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
values = np.random.randint(10, 100, size=len(categories))  # Random values for each category

# Step 2: Create a ColumnDataSource
source = ColumnDataSource(data=dict(categories=categories, values=values))

# Step 3: Create the bar chart figure
p = figure(x_range=categories, title="Random Categorical Bar Chart",
           toolbar_location=None, tools="")

# Step 4: Plot bars with color mapping based on values
p.vbar(x='categories', top='values', width=0.9, source=source,
       fill_color=factor_cmap('categories', palette=Viridis256, factors=categories))

# Step 5: Add hover tooltips to display the exact value
hover = HoverTool()
hover.tooltips = [("Category", "@categories"), ("Value", "@values")]
p.add_tools(hover)

# Step 6: Label the axes
p.xaxis.axis_label = "Categories"
p.yaxis.axis_label = "Values"

# Step 7: Remove gridlines for a cleaner look
p.xgrid.grid_line_color = None
p.y_range.start = 0

# Step 8: Display the chart
show(p)
```

```python
###20.Using Plotly, create a basic line plot of a randomly generated dataset, label the axes, and set the title as "Simple Line Plot"
```

```python
import plotly.graph_objects as go
import numpy as np

# Step 1: Generate random data
x = np.arange(0, 10, 0.1)  # X values from 0 to 10
y = np.random.randn(len(x))  # Random Y values

# Step 2: Create the plot
fig = go.Figure()

# Step 3: Add the line plot
fig.add_trace(go.Scatter(x=x, y=y, mode='lines', name='Random Line'))

# Step 4: Label the axes and set the title
fig.update_layout(title='Simple Line Plot', xaxis_title='X-axis', yaxis_title='Y-axis')

# Step 5: Show the plot
fig.show()
```

```python
###21. Using Plotly, create an interactive pie chart of randomly generated data, add labels and percentages, set the title as "Interactive Pie Chart"
```

```python
import plotly.graph_objects as go

# Step 1: Generate random data
labels = ['Category A', 'Category B', 'Category C', 'Category D']
values = np.random.randint(10, 100, size=4)  # Random values for each category

# Step 2: Create the pie chart
fig = go.Figure(data=[go.Pie(labels=labels, values=values, hole=.3)])

# Step 3: Set the title and show the percentages
fig.update_traces(hoverinfo='label+percent', textinfo='percent+label')

# Step 4: Set the title of the chart
fig.update_layout(title='Interactive Pie Chart')

# Step 5: Show the chart
fig.show()
```
